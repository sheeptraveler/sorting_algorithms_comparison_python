```python
import pandas as pd
from IPython.display import display

# Load and parse the data
data = []
with open('../metrics.txt', 'r') as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) >= 9 and parts[2] == 'orderedSet':
            data.append({
                'Algorithm': parts[0],
                'Size': int(parts[1]),
                'Time (s)': float(parts[8])
            })

# Create DataFrame
df = pd.DataFrame(data)

# Pivot to get algorithms as rows and sizes as columns
time_table = df.pivot_table(
    index='Algorithm',
    columns='Size',
    values='Time (s)',
    aggfunc='first'
).sort_index()

# Sort columns by size
time_table = time_table.reindex(sorted(time_table.columns), axis=1)

# Format and display the table
styled_table = (time_table.style
                .format('{:.6f}')
                .set_caption('Time Performance (seconds) on Ordered Data')
                .background_gradient(cmap='YlOrRd', axis=1)
                .highlight_min(axis=0, color='lightgreen'))

display(styled_table)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from cycler import cycler

# Set up the plot style
# plt.style.use('seaborn')
plt.figure(figsize=(12, 7))

# Define a custom color cycle for algorithms
colors = plt.cm.tab20.colors
custom_cycler = cycler(color=colors)
plt.gca().set_prop_cycle(custom_cycler)

# Load and filter the data
data = []
with open('../metrics.txt', 'r') as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) >= 9 and parts[2] == 'orderedSet':
            data.append({
                'Algorithm': parts[0],
                'Size': int(parts[1]),
                'Time (s)': float(parts[8])
            })

df = pd.DataFrame(data)

# Create the plot
for algorithm in df['Algorithm'].unique():
    algorithm_data = df[df['Algorithm'] == algorithm]
    plt.plot(algorithm_data['Size'], algorithm_data['Time (s)'], 
             marker='o', linestyle='-', linewidth=2, markersize=8, 
             label=algorithm)

# Customize the plot
plt.title('Performance on Ordered Data', fontsize=14)
plt.xlabel('Data Size', fontsize=12)
plt.ylabel('Time (seconds)', fontsize=12)
plt.yscale('log')  # Log scale to better see differences
plt.grid(True, which="both", ls="-")
plt.xticks(df['Size'].unique(), fontsize=10)
plt.yticks(fontsize=10)

# Add legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
